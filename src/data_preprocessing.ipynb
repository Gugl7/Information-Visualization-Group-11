# Data preprocessing for the ArtVis dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/artvis_dump_NEW-semicolon.csv', sep=';')
```

```python
print(df.head())
print("Form der Daten vor dem Filtern:", df.shape)

# Überprüfe die einzigartigen Jahre in 'e.startdate'
print("Einzigartige Werte in 'e.startdate':", df['e.startdate'].unique())


# Filtere nach Zeitraum 1905 bis 1910 und Geschlecht
filtered_df = df[(df['e.startdate'] >= 1905) & (df['e.startdate'] <= 1910) & 
                  (df['a.gender'].isin(['M', 'F']))]

# Überprüfe die Form der gefilterten Daten
print("Form der gefilterten Daten:", filtered_df.shape)

# Speichern der gefilterten Daten in eine neue CSV-Datei
filtered_df.to_csv('../data/processed_data.csv', index=False)

display(df)
print(df.shape)
```

```python
df.sort_values(by='e.id', inplace=True)
df.set_index('e.id', inplace=True)
```

```python
display(df.head())
print(df['e.type'].unique())
```

```python
import geopandas as gpd

import matplotlib.pyplot as plt

# Convert latitude and longitude to numeric values
df['e.latitude'] = pd.to_numeric(df['e.latitude'], errors='coerce')
df['e.longitude'] = pd.to_numeric(df['e.longitude'], errors='coerce')

# Drop rows with invalid latitude or longitude
df = df.dropna(subset=['e.latitude', 'e.longitude'])

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['e.longitude'], df['e.latitude']))

# Plot the positions on the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(figsize=(15, 10))
gdf.plot(ax=ax, color='red', markersize=5)
plt.show()
```

```python
df_filtered = df[(df['e.startdate'] >= 1905) & (df['e.startdate'] <= 1910)]
display(df_filtered)
```

```python
df_filtered = df_filtered[df_filtered['e.city'] != '-']
display(df_filtered)
print(df_filtered['e.city'].unique())
```

```python
df_filtered.to_csv('../data/artvis-filtered.csv', sep=';', index=False)
```
